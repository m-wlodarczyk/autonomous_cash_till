```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, classification_report, mean_squared_error, mean_absolute_error
%matplotlib inline
```

```python
y_pred = np.load("Test/y_pred_70_score_80_iou.npy")
y_true = np.load("Test/y_true_70_score_80_iou.npy")
```

```python
cm = confusion_matrix(y_true, y_pred)
report = classification_report(y_true, y_pred, target_names=["not found", 'gummy', 'peanuts', 'snickers', 'toffifee', 'knoppers', 'chocolate', 'cookies', 'corn_cakes', 'bubble_gum', 'lollipop', 'chips'])

dic = classification_report(y_true, y_pred, output_dict=True)
mse = mean_squared_error(y_true, y_pred)
mae = mean_absolute_error(y_true, y_pred)
df = pd.DataFrame(dic).transpose()
```

```python
print(report)
```

```python
print("MSE: ", mse)
```

```python
print("MAE: ", mae)
```

```python
fig =plt.figure(figsize=(14,8))

products = ["not found", 'gummy', 'peanuts', 'snickers', 'toffifee', 'knoppers', 'chocolate','cookies', 'corn_cakes', 'bubble_gum', 'lollipop', 'chips']
ax = sns.heatmap(cm,annot=True,cbar=False, linewidths=.5, annot_kws={"size": 18}, cmap='coolwarm', fmt='g' )
ax.set_xlabel("Predicted label", fontsize = 15)
ax.set_ylabel("Actual label", fontsize = 15)
ax.set_xticklabels( products, fontsize = 12, rotation=30)
ax.set_yticklabels( products, fontsize = 15, rotation=0)
fig.savefig("Test/confusion_matrix_70_score_80_iou.png")
```
